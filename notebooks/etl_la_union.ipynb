```python
paths = set_paths('union')
csvs = [p for p in paths['inputs'].glob('LaUnion_Juvenile_Adults.csv')]
csv = csvs[0]
data = pd.read_csv(csv)
data
```

```python
shorthands = {'hysi': 'hyparrhenia_sinaica', 'pimi': 'piptatherum_miliaceum',
                'osla': 'osyris_lanceolata', 'saop': 'salsola_oppositifolia',
                'stte': 'stipa_tenacissima', 'atha': 'atriplex_halimus',
                'hest': 'helichrysum_stoechas', 'taca': 'tamarix_canariensis',
                'lysp': 'lygeum_spartum', 'piha': 'pinus_halepensis',
                'pasu': 'paronychia_suffruticosa', 'dope': 'dorycnium_pentaphyllum',
                'lica': 'limonium_carthaginense'}

df = data.copy()

df['nurse'] = df.nurse.replace(shorthands)

species = df.columns[-170:].tolist()
# nurses_sorted = density.nurse.value_counts(ascending=False).index.tolist()
# juveniles_sorted = density[species].mean(axis=0).sort_values(ascending=False).index.tolist()
# density.nurse = density.nurse.astype("category").cat.set_categories(nurses_sorted)
# density.insert(17, 'area', math.pi/4 * density.dia1 * density.dia2 / 10000)

df[species] = df[species].gt(0)
df
```

```python
# Include nurse into each patch, considering it a normal plant like the others
for sp in df.nurse.unique():
    df.loc[(df.nurse == sp) & (df.microhabitat == 'parche'), sp] = True

df
```
